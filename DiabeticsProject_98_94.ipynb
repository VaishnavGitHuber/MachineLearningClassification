```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Dataset of Diabetes .csv")
df = data.copy()
df.head()
```

```python
df.shape
```

* Removing noises in the target features

```python
df["CLASS"].unique()
```

```python
# I am removing the features with class label is not "Y" and "N"
df = df[df["CLASS"].isin(["Y","N"])]
df.shape
```

### Removing the constant features
* ID
* No_Pation

```python
df.drop(columns=["ID","No_Pation"],inplace=True)
df.head(2)
```

```python
df["Gender"].value_counts()
```

```python
df["Gender"][df["Gender"] == "f"] = "F"
```

```python
df["Gender"].value_counts()
```

```python
df["CLASS"].value_counts()
```

* All noises in the catagorical features are now cleared

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
features_to_encode = ["CLASS","Gender"]
for features in features_to_encode:
  df[features] = encoder.fit_transform(df[features])
df.head()
```

* All features are encoded now its time for splitting the dataset in training and testing dataset

```python
X = df.drop(columns=["CLASS"])
Y = df["CLASS"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=42,shuffle=True)
```

```python
X_test.shape,X_train.shape
```

```python
Y_train.shape,Y_test.shape
```

```python
# Using Random forest classfier since the dataset is imbalaced
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score

# training model
model = RandomForestClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

# Accuracy score
acuracy_score = accuracy_score(y_pred,Y_test)
print("Accuracy scaore: ",acuracy_score)

# confusion matrics
matrix = confusion_matrix(Y_test,y_pred)
print(matrix)
```
