```python
# importing the necessory modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train.csv")
df = train_df.copy()
```

```python
df.head()
```

```python
plt.title("Missing value heatmap")
sns.heatmap(df.isnull())
plt.show()
```

from this heat map we understood,
- Majority of the data points in Cabin feature is missing,some of the age category is also is missing

```python
plt.title("Missing value: Bar plot")
df.isnull().mean().plot(kind = "bar",color="r")
```

```python
df.shape
```

* There is 891 data entries and about 12 features

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
df["Sex"].value_counts().plot(kind="bar",title="The Sex: bar plot",color="k",edgecolor="k",width=0.2)
plt.subplot(1,2,2)
df.groupby("Sex")["Survived"].sum().plot(kind="bar",color="r",width=0.2,title="Sex vs Survival chance")
plt.show()
```

* eventhough the travellers are more of males but females have a greater chance of survival so sex gives more information for the prediction

```python
df["Sex"].map({"male":1,"female":2}).corr(df["Survived"]) # high corelation
```

```python
# so, females have the greater chance of survival
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
df["Pclass"].value_counts().plot(kind="bar",title="The passenger class: bar plot",color="m",edgecolor="k",width=0.2)
plt.subplot(1,2,2)
df.groupby("Pclass")["Survived"].sum().plot(kind="bar",color="r",width=0.2,title="Passenger calss vs Survival")
plt.show()
```

```python
# passenger class with 1 have the greater chamnce of survival
# passengers are more in class 3 but the passengers in class 1 have high survival rate
df["Pclass"].corr(df["Survived"])
```
