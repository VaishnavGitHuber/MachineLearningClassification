```python
# A normal result is 0.7 to 1.3 mg/dL (61.9 to 114.9 µmol/L) for men and 0.6 to 1.1 mg/dL (53 to 97.2 µmol/L)
# for women
print("Normal Cretin level for male is 61.9 - 114.9 and for females 53 - 97.2")
print("The range of Creatin level in the dataset is {} to {} and mean for this {}".format(df["Cr"].min(),df["Cr"].max(),df["Cr"].mean()))
```

```python
dataset = df[["Gender","Cr"]].copy()
```

```python
x = df[["Gender","Cr"]][df['Gender'] == "M"]
y = df[["Gender","Cr"]][df["Gender"] == "F"]
```

```python
plt.figure(figsize=(10,6))
plt.title("Cretin ratio Gender Scatter plot")
plt.scatter(x.index,x["Cr"],color="r",label="Male")
plt.scatter(y.index,y["Cr"],color="k",label="Female")
plt.legend(loc="upper left")
plt.xlabel("<--------------Data points------------>")
plt.ylabel("<--------------Creatin level---------->")
plt.grid(True,color="g")
plt.show()
```

* A significant number of outlier datapoints can be observed with in famale category on the scatter plot
* let us check how many of the male and females belongs to the categery of the safe and how many of them belongs to the category of unsafe

```python
# Count of percent of Males in the category of safe and no of people in unsafe
# safe range (61.9 to 114.9 µmol/L)
safe_count = 0
unsafe_count = 0
for i in x["Cr"]:
  if i > 61.9 and i < 114.9:
    safe_count += 1
  else:
    unsafe_count += 1
print("IN MALES: No of safe percent: {}% and Unsafe percent {}%".format(100 * round(safe_count/len(x["Cr"]),3),100* round(unsafe_count/len(x["Cr"]),3)))
```

```python
# Count of percent of females in the category of safe and no of people in unsafe
# safe range (53 to 97.2 µmol/L)
safe_count = 0
unsafe_count = 0
for i in y["Cr"]:
  if i > 53 and i < 97.2:
    safe_count += 1
  else:
    unsafe_count += 1
print("IN MALES: No of safe percent: {}% and Unsafe percent {}%".format(100*round(safe_count/len(y["Cr"]),3), 100*round(unsafe_count/len(y["Cr"]),3)))
```
